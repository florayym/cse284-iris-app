```python
#get samples names 
import pandas as pd
samples_list = pd.read_csv('samples_ID.txt', sep='\t', header=None, error_bad_lines=False)
samples_list = samples_list.T[0]
```

```python
SNPs_total = len(set(all_GT['rsid']))
number_samples = len(samples_list)

start_idx = 0
for i in range(0,SNPs_total):
    c = 0 
    for b in range(0+start_idx,len(samples_list)+start_idx):
        all_GT['sample'].iloc[b] = samples_list[c]
        c = c+1 
    start_idx = start_idx + number_samples

    
#translate genotypes into 0,1,2 information
value = []
for i in range(len(all_GT)):  
    rsid = all_GT['rsid'].iloc[i]  #get rsid 
    minor_allele = bed_file[bed_file['rsid'] == rsid].iloc[0,2] #find corresponding minor allele
    if all_GT['GT'].iloc[i] == '1|1':
        if minor_allele == all_GT['ALT'].iloc[i]:
            value.append(2)
        else:
            value.append(0)
    elif all_GT['GT'].iloc[i] == '0|1':
        if minor_allele == all_GT['ALT'].iloc[i] or minor_allele == all_GT['REF'].iloc[i]:
            value.append(1)
        else:
            value.append(0)
    elif all_GT['GT'].iloc[i] == '1|0':
        if minor_allele == all_GT['ALT'].iloc[i] or minor_allele == all_GT['REF'].iloc[i]:
            value.append(1)
        else:
            value.append(0)
    elif all_GT['GT'].iloc[i] == '0|0':
        if minor_allele == all_GT['REF'].iloc[i]:
            value.append(2)
        else:
            value.append(0)
        
all_GT['X'] = value
```

```python
import math 

alpha1 = 1.70
alpha2 = -1.26
alpha3 = -2.62

results = pd.DataFrame(columns = ['pBlond', 'pBrown', 'pRed', 'pBlack'], index = samples_list,dtype=float)

for sample in samples_list: #for each sample
    
    table =  all_GT[all_GT['sample'] == sample]  #only select values for that sample 
    sum_b1 = 0
    sum_b2 = 0
    sum_b3 = 0
    rsid_R = ['rs1805006','rs11547464','rs1805007','rs1805008','rs1805009']
    rsid_r = ['rs1805005','rs2228479','rs1110400','rs885479']
    X_R = 0
    X_r = 0

    #first calculate a cumulative X for the MC1R_R and MC1R_r
    count_hom = 0 
    count_heter = 0
    count_hom_alt = 0
    for rsid in rsid_R:
        if table[table['rsid'] == rsid].X.iloc[0] == 0:
            count_hom = count_hom + 1
        elif table[table['rsid'] == rsid].X.iloc[0] == 1:
            count_heter = count_heter+1
        else: #if X = 2 
            count_hom_alt = count_hom_alt +1
            
    if count_hom == len(rsid_r):
        X_R = 0
    elif count_hom_alt >= 1:
        X_R = 2
    elif count_heter == 1:
        X_R = 1
    else:
        X_R = 2
        
    #first calculate a cumulative X for the MC1R_R and MC1R_r
    count_hom = 0 
    count_heter = 0
    count_hom_alt = 0
    for rsid in rsid_r:
        if table[table['rsid'] == rsid].X.iloc[0] == 0:
            count_hom = count_hom +1
        elif table[table['rsid'] == rsid].X.iloc[0]  == 1:
            count_heter = count_heter+1
        else: #if X = 2 
            count_hom_alt = count_hom_alt +1
            
    if count_hom == len(rsid_r):
        X_r = 0
    elif count_hom_alt >= 1:
        X_r = 2
    elif count_heter == 1:
        X_r = 1
    else:
        X_r = 2

    for i,rid in enumerate(bed_file_concentrated['rsid']):  #go through SNPs
    # I would add MC1R_R & MC1R_r to the bed file 
    # in here I would probably add an if statement for MC1R_R & MC1R_r
    # and then would iterate through the list of rsids for each of those 
    # and decide value of X based on that 
    # would probably make a series of if statements (basically if you find homozygous, X = 2, etc)
        if rid != 'MC1R_R' and rid != 'MC1R_r':
            rsid = bed_file_concentrated['rsid'].iloc[i]
            beta1 = bed_file_concentrated['𝛽1'].iloc[i]
            beta2 = bed_file_concentrated['𝛽2'].iloc[i]
            beta3 = bed_file_concentrated['𝛽3'].iloc[i]
            X = table[table['rsid'] == rid].X.iloc[0]
            sum_b1 = sum_b1 + beta1 * X
            sum_b2 = sum_b2 + beta2 * X
            sum_b3 = sum_b3 + beta3 * X
        elif i == 'MC1R_R':
            rsid = bed_file_concentrated['rsid'].iloc[i]
            beta1 = bed_file_concentrated['𝛽1'].iloc[i]
            beta2 = bed_file_concentrated['𝛽2'].iloc[i]
            beta3 = bed_file_concentrated['𝛽3'].iloc[i]
            X = X_R #calculated before
            sum_b1 = sum_b1 + beta1 * X
            sum_b2 = sum_b2 + beta2 * X
            sum_b3 = sum_b3 + beta3 * X
        elif i == 'MC1R_r':
            rsid = bed_file_concentrated['rsid'].iloc[i]
            beta1 = bed_file_concentrated['𝛽1'].iloc[i]
            beta2 = bed_file_concentrated['𝛽2'].iloc[i]
            beta3 = bed_file_concentrated['𝛽3'].iloc[i]
            X = X_r #calculated before
            sum_b1 = sum_b1 + beta1 * X
            sum_b2 = sum_b2 + beta2 * X
            sum_b3 = sum_b3 + beta3 * X
            
    results.loc[sample, 'pBlond'] = (math.exp(alpha1+sum_b1))/(1+math.exp(alpha1+sum_b1)+math.exp(alpha2+sum_b2)+math.exp(alpha3+sum_b3))
    results.loc[sample, 'pBrown'] = (math.exp(alpha2+sum_b2))/(1+math.exp(alpha1+sum_b1)+math.exp(alpha2+sum_b2)+math.exp(alpha3+sum_b3))
    results.loc[sample, 'pRed'] = (math.exp(alpha3+sum_b3))/(1+math.exp(alpha1+sum_b1)+math.exp(alpha2+sum_b2)+math.exp(alpha3+sum_b3))
    results.loc[sample, 'pBlack'] = 1 - results.loc[sample, 'pBlond'] - results.loc[sample, 'pBrown'] - results.loc[sample, 'pRed']
    
#the results dataframe contains all the hair color probabilities for all samples with 
#the corresponding ancestry information

results.insert(0, 'population',sample_pop_match)
results = results.dropna()
results.groupby(['population']).mean()
```
